```python
#Problem 1: Unique Words in a Sentence
sentence = input("Enter Your Sentence: ")

# Splitting sentence into a list of words
words = sentence.split()

# Using a set to automatically remove duplicates
unique_words = set(words)

# Displaying count
print("Unique words count:", len(unique_words))

# Displaying the unique set
print("Unique words:", unique_words)
     
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Using max() with a key to find the employee name with the highest value
highest_paid = max(employees, key=employees.get)

print(f"Highest Salary: {highest_paid} - {employees[highest_paid]}")
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]

# Initializing with the first element
max_val = numbers[0]
min_val = numbers[0]

# Iterating through the list to compare values
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count = 0

for p in prices:
    if p > 1000: # Comparing each price with 1000
        count += 1

print("Products above 1000:", count)
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P") # Counting present days
total_days = len(attendance) # Getting total days

percentage = (present_count / total_days) * 100 # Calculating percentage
print(f"Attendance Percentage: {percentage}")
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers) # Sets remove duplicates automatically

print("Unique phone numbers:", unique_numbers)
```

```python
#Problem 7: Count Character Frequency
text = input("Enter your string : ")
freq = {}

for char in text:
    # Incrementing count for each character occurrence
    freq[char] = freq.get(char, 0) + 1

print(freq)
```

```python
#Problem 8: Convert List to Tuple
data_list = [10, 20, 30]
converted_tuple = tuple(data_list)

print("Tuple:", converted_tuple)
```

```python
#Problem 9: Check if a Key Exists in a Dictionary

employee_data = {"Name": "Ravi", "ID": 101}
search_key = "Name"

if search_key in employee_data: # Using the 'in' keyword
    print("Employee exists")
```

```python
#Problem 10: Calculate Average Marks

marks = [80, 75, 60, 75]

total_sum = sum(marks) # Summing all values
average = total_sum / len(marks) # Dividing by total number of marks

print("Average Marks:", average)
```

**Task 1: Replace Hardcoded Prompts**

```python
from langchain_core.prompts import PromptTemplate

# Task 1 : Reolacing Hardcoded prompt to Langchain prompt
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(explain_template.format(topic="Machine Learning"))
```

**Task 2: Multi-Input Prompt System**

```python
# Task 2 : Multi-Input Prompt System
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

tests = [
    ("AI", "beginners", "friendly"),
    ("Python", "kids", "fun"),
    ("Deep Learning", "engineers", "technical")
]

for t in tests:
    print(multi_template.format(topic=t[0], audience=t[1], tone=t[2]))
```

**Task 3: Prompt Variations Engine**

```python
# Task 3: Prompt Variations Engine

teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_prompt.format(topic=topic))
print(interview_prompt.format(topic=topic))
print(story_prompt.format(topic=topic))
```

**Task 4: ChatPromptTemplate System**

```python
from langchain_core.prompts import ChatPromptTemplate

# Task 4: ChatPromptTemplate System
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role}."),
    ("user", "Explain {topic}")
])

print(chat_template.format(role="teacher", topic="Neural Networks"))
print(chat_template.format(role="interviewer", topic="Neural Networks"))
print(chat_template.format(role="motivator", topic="Neural Networks"))
```

**Task 5: Input Validation Layer**

```python
# Task 5 : validation
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        raise ValueError("Invalid audience type")

    if tone not in valid_tone:
        raise ValueError("Invalid tone type")

    return audience, tone
```

**Task 6: Prompt Generator App**

```python
# Task 6 : Prompt Generator
def generate_prompt(topic, audience, tone, style):

    validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} clearly step by step for {audience} in a {tone} tone",
        "interview": "Ask 3 questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
    }

    if style not in styles:
        raise ValueError("Invalid style")

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=styles[style]
    )

    return template.format(topic=topic, audience=audience, tone=tone)

```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7: Template Reusability Test**

```python
# Task 7 : Template Reusablity
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Give a short introduction to {topic}"
)

topics = ["AI", "Data Science", "Cybersecurity", "Cloud Computing", "Blockchain"]

for t in topics:
    print(reuse_template.format(topic=t))
```

**PipeLine Flow**

```python
user_input = ("Deep Learning", "beginner", "fun", "teaching")

validated = validate_inputs(user_input[1], user_input[2])
output = generate_prompt(user_input[0], validated[0], validated[1], user_input[3])

print(output)
```

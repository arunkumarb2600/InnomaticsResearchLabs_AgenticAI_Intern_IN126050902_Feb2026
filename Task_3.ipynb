```python
#Problem 1: Employee Performance Bonus Eligibility
# Dictionary of employees and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
highest_score = max(employees.values())

# Find employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Problem 2: Search Query Keyword Analysis
import string

# Input query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = query.split()

# Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Filter words appearing more than once
repeated_words = {word: count for word, count in frequency.items() if count > 1}

print(repeated_words)
```

```python
#Problem 3: Sensor Data Validation
# Sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid readings (even numbers) with their index
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem 4: Email Domain Usage Analysis
# List of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains and count usage
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

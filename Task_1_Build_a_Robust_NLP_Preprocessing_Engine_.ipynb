### **1.What is the difference between "Love" and "love" in NLP?**
In NLP, "Love" and "love" are treated as different words because of capitalization.
"Love" can be a name or start of a sentence, while "love" means emotion.
So, we usually convert all words to lowercase to avoid confusion.

### **2.What happens if stopwords are not removed?**
If stopwords (like is, the, and) are not removed:
Data becomes large and noisy,
Important words get less focus,
Model performance may reduce.

### **3.Provide two real-world scenarios where removing stopwords can be harmful.**
Sentiment Analysis
"I do not like this" → becomes "like" (wrong meaning)
Question Systems.

"What is your name?" → becomes "name" (loses meaning)

### **4.Explain the difference between stemming and lemmatization.**
Stemming: Cuts words roughly

Example: studies → studi

Lemmatization: Gives correct base word

Example: studies → study

Stemming is faster, Lemmatization is more accurate.



# **Task 2:Build Advanced Preprocessing Function**


```python
import re

def preprocess_text(text):

    text = text.lower()

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)

    # Fix: keep 2 letters (goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return " ".join(words)
```

```python
text = "WOW!!! This is soooo goooood!!! Visit http://example.com now, I have 2 dogs!!!"
print(preprocess_text(text))
```

# **Task 4: Token Analytics**


```python
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print("Original Text :", text)
    print("Cleaned Sentence:", cleaned)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-" * 50)
```

# **Task 5: Frequency Analysis**


```python
from collections import Counter

texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

# collect all tokens
for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

# count frequency
counter = Counter(all_tokens)

# Top 10 most frequent words
most_common = counter.most_common(10)

# Top 5 least frequent words
least_common = counter.most_common()[-5:]

print("Top 10 Most Frequent Words:")
for word, count in most_common:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_common:
    print(word, ":", count)
```

# **Task 6: Build Full Pipeline**


```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        cleaned = preprocess_text(text)   # use your previous function
        tokens = cleaned.split()

        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product"
]

result = full_pipeline(texts)

print(result)
```

# **Task 7: Error Handling**


```python
import re

def preprocess_text(text):

    # Handle empty input
    if not text or not isinstance(text, str):
        return ""

    text = text.lower()

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return " ".join(words)
```

```python
test_cases = [
    "",                # empty string
    "😍😂🔥",       # only emojis
    "1234567890"       # only numbers
]

for text in test_cases:
    print("Original:", text)
    print("Processed:", preprocess_text(text))
    print("-" * 30)
```

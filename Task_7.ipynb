```python
#Question 1 - Smart Light Controller

# Class to control smart light
class SmartLight:
    # Constructor to initiate name and set default status 
    def __init__(self, name):
        self.name = name
        self.status = "OFF" 

    # Method to turn on the light
    def turn_on(self):
        self.status = "ON"

    # Method to turn off the light 
    def turn_off(self):
        self.status = "OFF"

    # method to display the status
    def display_status(self):
        print(f"{self.name}  is {self.status}")

# Getting the user inputs 
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").strip().upper()

# Creating the object of smartlight class
light = SmartLight(light_name)

# Performing the action based on the input given action: ON/OFF
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action!")

# printing the status
light.display_status()
```

```python
#Question 2 - Employee ID Card System
# Class employee to handle employee data 
class Employee:

    # constructor to store the details of the employee
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    # Function to display employee data
    def show_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


# creating object for the class employee
emp1 = Employee("Rahul", "EMP102", "IT")

# Calling the function using object
emp1.show_id_card()
```

```python
#Question 3 - Mobile Contact Record
# Class to handle mobile contacts 
class Contact:

    # constructor to store details of the mobile contact
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone   

    # function to display contact information
    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# creating an object for class Contact
c1 = Contact("Anita", "9876543210")

# calling the function using an object c1
c1.show_contact()
```

```python
#Question 4 - Product Price Tag Generator
# Class to generate a product price tag
class Product:

    # constructor to store the details of the product
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # function to print the price of the product
    def show_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# creating an odject for class Product
item = Product("Headphones", 2499)

# Calling the function using an objet item
item.show_price_tag()
```

```python
#Question 5 - Movie Rating Display System
# class to handle movie ratings
class Movie:

    # constructor to store the details of the movie and rating
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating 

    # function to print the name and rating
    def show_details(self):
        print("Movie:", self.name)
        print("Rating:", str(self.rating) + " / 5")


# creating an object for class Movie 
m1 = Movie("Inception", 4.8)

# calling the show_details function using object m1
m1.show_details()
```

```python
#Question 6 - Delivery Address Manager
# class to handle delivery address
class Delivery:

    # constructor to store the details of the customer
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # function to print the customer name and address 
    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# creating object foe class Delivery
order1 = Delivery("Suman", "Hyderabad")

# calling the function show_details using object order1
order1.show_details()
```

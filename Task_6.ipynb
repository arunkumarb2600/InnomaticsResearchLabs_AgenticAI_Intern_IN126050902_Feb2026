```python
# Logic_Building_Task_6:

# Problem 1: Smart Parking Lot Management System

def parking_system(capacity, logs):
    """
    This function calculates current parked vehicles
    and checks if parking is available or full.
    """

    parked = 0
    max_used = 0

    # Count IN and OUT logs
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT":
            parked -= 1

        # Track peak usage
        if parked > max_used:
            max_used = parked

    # Check capacity
    if parked > capacity:
        status = "Full"
    else:
        status = "Available"

    return parked, status


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

current, parking_status = parking_system(capacity, vehicle_logs)

print("Currently Parked Vehicles:", current)
print("Parking Status:", parking_status)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery(distance, traffic, weather):
    """
    Calculates estimated delivery time
    based on distance, traffic and weather.
    """

    base_time = distance * 5   # 5 mins per km

    # Add delay for traffic
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5

    # Add delay for weather
    if weather == "Rainy":
        base_time += 5

    return base_time


distance = 8
traffic_level = "High"
weather_condition = "Rainy"

eta = estimate_delivery(distance, traffic_level, weather_condition)

print("Estimated Delivery Time:", eta, "minutes")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def check_occupancy(total_seats, booked_seats):
    """
    Calculates occupancy percentage
    and checks show status.
    """

    booked_count = len(booked_seats)

    percentage = (booked_count / total_seats) * 100

    if percentage == 100:
        status = "Housefull"
    elif percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return int(percentage), status


total = 200
booked = [1] * 150   # 150 seats booked

occ, show_status = check_occupancy(total, booked)

print("Occupancy:", str(occ) + "%")
print("Show Status:", show_status)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    """
    This function calculates the average CPU load
    and classifies the server status.
    """

    total = 0

    # Calculate total CPU usage
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    # Classification logic
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return int(average), status


cpu_data = [45, 60, 70, 85, 90]

avg_load, server_status = classify_server_load(cpu_data)

print("Average CPU Load:", str(avg_load) + "%")
print("Server Status:", server_status)
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def check_resource_usage(resource_dict):
    """
    This function checks if any resource is overused.
    Usage above 8 hours is considered overuse.
    """

    overused = []

    for resource, hours in resource_dict.items():
        if hours > 8:
            overused.append(resource)

    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    return overused, alert


resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, energy_alert = check_resource_usage(resources)

print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", energy_alert)
     
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, registrations):
    """
    This function checks confirmed and waitlisted users
    based on event capacity.
    """

    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    return confirmed, waitlisted, status


event_capacity = 100
total_registrations = 105

confirmed_users, waitlist_users, registration_status = manage_registrations(event_capacity, total_registrations)

print("Confirmed Registrations:", confirmed_users)
print("Waitlisted Users:", waitlist_users)
print("Registration Status:", registration_status)
```
